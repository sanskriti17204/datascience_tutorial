Label encoding

converting the labels into numeric form

```python
#importing the dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label encoding of breast cancer dataset

```python
#loading the data from csv file to pandas data frame
cancerdata= pd.read_csv('/content/breast cancer.csv')
```

```python
cancerdata.head()
```

```python
cancerdata['diagnosis'].value_counts()
```

```python
#load the label encoder function
labelencode=LabelEncoder()
```

```python
label=labelencode.fit_transform(cancerdata.diagnosis)
```

```python
#appending yhe labels to the dataframe
cancerdata['target']=label
```
