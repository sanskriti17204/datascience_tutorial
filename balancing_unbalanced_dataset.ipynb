```python
import pandas as pd
import numpy as np
```

```python
credit=pd.read_csv('/content/creditcard.csv')
```

```python
credit.head()
```

```python
credit['Class'].value_counts()
```

this is highly unbalanced dataset

0-> legit transactions
1-> fraudelent transactions

```python
#sepearting the legit and fraudelent transactions
legit=credit[credit.Class==0.0]
fake=credit[credit.Class==1.0]
```

```python
print(legit.shape)
print(fake.shape)
```

**Under sampling**

build a simple dataset containing distributiom of fraudelent transactions

for eg a dataset containing 38 fake and 38 legit transactions

```python
legit_sample=legit.sample(n=38)
```

```python
print(legit_sample.shape)
```

Concatenate the two dataframe

```python
new_dataset=pd.concat([legit_sample, fake], axis=0)
```

```python
new_dataset.head()
```
