```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,100)  #generating 100 valyes between 1 and 10
y=np.sin(x)
z=np.cos(x)  # z will take values of cos x for all values of x
plt.plot(x,y)  #plotting x vs y
plt.plot(x,z)  #plotting y vs z
plt.show()  #mandatory to make the graph visible
```

```python
plt.plot(x,z,'g*')
plt.xlabel('angle')
plt.ylabel('sine values')
plt.title('sine wave')
plt.show()
```

```python
x=np.linspace(-10,10,20)  #for parabola
q=x**2
plt.plot(x,q,'r_')   #r stands for red and _ is the style type
plt.show()
```

```python
#plot a bar plot
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
food=['pizza','noodles','rice','curry']
children=[10,50,5,70]
ax.bar(food,children)
plt.xlabel('food')
plt.ylabel('people')
plt.show()
```

```python
#plot a pie chart
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
food=['pizza','noodles','rice','curry']
children=[10,50,5,70]
ax.pie(children,labels=food,autopct='%0.2f%%')
plt.show()
```

```python
#scatter plot
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
#plt.show()
```

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,cmap='Blue')
plt.show()
```
