```python
import seaborn as sns  #has built in datasets
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
tip=sns.load_dataset('tips')
tip.head()
```

```python
#visualize the data
sns.relplot(data=tip,x='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size')
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
#scatter plot
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris)
```

```python
sns.relplot(x='sepal_length', y='petal_length', hue='species', data=iris)
```

```python
sns.set_theme()  #setting a theme for the plots. wiull show grid
```

```python
#loading titanic dataset
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
print(sns.countplot(x='class', data=titanic))
```

```python
sns.countplot(x='survived', data=titanic)
```

```python
sns.barplot(x='sex', y='survived', hue='class', data=titanic)
```

```python
sns.displot(iris['sepal_length'])
```

```python
sns.distplot(iris['sepal_length'])
```

CORRELATION
1- POSITIVE CORRELATION(increase with increase)
2- NEGATIVE CORRELATION(decrease with increase)

```python
correlation=iris.corr()
```

```python
#constructing a heat map
plt.figure(figsize=(10,10))
data=sns.heatmap(correlation, cbar=True, square=True, fmt='.if', annot=True, annot_kws={'size':8}, cmap='Blues')
plt.imshow( data )
```
